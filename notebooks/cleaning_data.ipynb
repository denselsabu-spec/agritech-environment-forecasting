```python
import pandas as pd
df = pd.read_csv("../data/mushroom_yield_raw.csv")
print(df.head())
```

```python
#to check missing values
print("\n Missing Values: ")
print(df.isnull().sum())
```

```python
#Replace the empty columns with median values
numeric_columns = ["temperature_c","humidity_percent","co2_ppm","yield_kg"]
for column in numeric_columns:
    df[column] = df[column].fillna(df[column].median())
```

```python
#Remove duplicates rows
df.drop_duplicates(inplace=True)
```

```python
#Reset Index after cleaning
df.reset_index(drop=True,inplace=True)
```

```python
#the cleaned data
print("\n Missing Values: ")
print(df.isnull().sum())

print("\nCleaned data")
print(df.head())
```

```python
df.to_csv("../data/mushroom_yield_cleaned.csv",index=False)
```
