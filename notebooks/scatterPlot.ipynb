```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the dataset
df = pd.read_csv("../data/mushroom_yield_cleaned.csv")
```

```python
#Scatter Plots
#temperature vs Yield
plt.figure(figsize=(6,4))
sns.scatterplot(x="temperature_c", y = "yield_kg", data = df)
plt.title("Temperature vs Yield")
plt.show()
```

```python
#Scatter Plots
#Humidity vs Yield
plt.figure(figsize=(6,4))
sns.scatterplot(x="humidity_percent", y = "yield_kg", data = df)
plt.title("Humidity vs Yield")
plt.show()
```

```python
#Scatter Plots
#CO2 vs Yield
plt.figure(figsize=(6,4))
sns.scatterplot(x="co2_ppm", y = "yield_kg", data = df)
plt.title("CO2 vs Yield")
plt.show()
```
